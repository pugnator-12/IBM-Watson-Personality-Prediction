Tried comparative study of few personalities
Studied variance in traits based on extracts from autobiographies

```python
import json
import pandas as pd
from matplotlib import pyplot as plt 
import seaborn as sns
from ibm_watson import PersonalityInsightsV3
url ='https://gateway-lon.watsonplatform.net/personality-insights/api'
apikey = '<enter you API key>'
service = PersonalityInsightsV3(url=url, version='2017-10-13', iam_apikey=apikey)
```

```python
string = "extract0.txt"
label = ['Nelson_Mandela','Michelle Obama','Malala']
openness = []
Conscientiousness = []
Extraversion  = []
Agreeableness = []
Emotional_range = []
for i in range (1,4):
    x = i;
    no = str(i);
    name = string.replace("0",no)
    f = open(name,"r")
    text = f.read()
    profile = service.profile(text, accept='application/json').get_result()
    pts = profile['personality']
    dict_pt = {pt['name']:pt['percentile'] for pt in pts}
    df1 = pd.DataFrame.from_dict(dict_pt, orient='index')
    df1.reset_index(inplace=True)
    df1.columns = ['trait', 'percentile']
    openness.append(df1.iat[0,1])
    Conscientiousness.append(df1.iat[1,1])
    Extraversion.append(df1.iat[2,1])
    Agreeableness.append(df1.iat[3,1])
    Emotional_range.append(df1.iat[4,1])
```

```python
import numpy as np
#openness_array = np.asarray(openness)
#Conscientiousness_array = np.asarray(Conscientiousness)
#Extraversion = np.asarray(Extraversion)
#Agreeableness_array = np.asarray(Agreeableness)
#Emotional_range_array = np.asarray(Emotional_range)

index = np.arange(len(label))
```

```python
plt.bar(index, openness)
plt.xlabel('openness', fontsize=10)
plt.ylabel('percentile', fontsize=10)
plt.xticks(index, label, fontsize=10, rotation=30)
plt.show()

plt.bar(index, Conscientiousness)
plt.xlabel('Conscientiousness', fontsize=10)
plt.ylabel('percentile', fontsize=10)
plt.xticks(index, label, fontsize=10, rotation=30)
plt.show()

plt.bar(index, Extraversion)
plt.xlabel('Extraversion', fontsize=10)
plt.ylabel('percentile', fontsize=10)
plt.xticks(index, label, fontsize=10, rotation=30)
plt.show()

plt.bar(index, Agreeableness)
plt.xlabel('Agreeableness', fontsize=10)
plt.ylabel('percentile', fontsize=10)
plt.xticks(index, label, fontsize=10, rotation=30)
plt.show()

plt.bar(index, Emotional_range)
plt.xlabel('Emotional_range', fontsize=10)
plt.ylabel('percentile', fontsize=10)
plt.xticks(index, label, fontsize=10, rotation=30)
plt.show()
```

```python
N = 5
ind = np.arange(N)  # the x locations for the groups
width = 0.27       # the width of the bars

fig = plt.figure()
ax = fig.add_subplot(111)

yvals = [openness[0],Conscientiousness[0],Extraversion[0],Agreeableness[0],Emotional_range[0]]
rects1 = ax.bar(ind, yvals, width, color='r')
zvals = [openness[1],Conscientiousness[1],Extraversion[1],Agreeableness[1],Emotional_range[1]]
rects2 = ax.bar(ind+width, zvals, width, color='g')
kvals = [openness[2],Conscientiousness[2],Extraversion[2],Agreeableness[2],Emotional_range[2]]
rects3 = ax.bar(ind+width*2, kvals, width, color='b')

ax.set_ylabel('Percentile')
ax.set_xticks(ind+width)
ax.set_xticklabels( ('openness', 'Conscientiousness', 'Extraversion','Agreeableness','Emotional_range'),rotation=30 )
ax.legend( (rects1[0], rects2[0], rects3[0]), ('Nelson Mandela', 'Michelle Obama', 'Malala') )


plt.show()
```
